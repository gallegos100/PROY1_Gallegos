```python
import pandas as pd
from pandas import Series

print("¿Cuales fueron las ventas de los ultimos 5 dias?")

lista1 = [["Pan", "Mantequilla", "Mermelada", "Jamonada","Mayonesa"],
          [5,7,4,7,8],
          [0.30, 1.5,1.80,1.20,2.50],
          [0,0,0,0,0]
         ]

df1 = pd.DataFrame(lista1)
    
df1.columns=["Lunes", "Martes", "Miercoles", "Jueves", "Viernes"]

df1.index = Series (["Producto", "# Ventas", "Precio S/.", "Descuento",])


df2 = df1.transpose()

df2["Descuento"] = 0.95*df2["Precio S/."]

df2
```

EJERCICIO N° 2

```python
Ingresos = {
    "Luis": 3500,
    "Harold": 2100,
    "Carlitos": 1800,
    "Juanca": 2000,  
    }
```

```python
import pandas as pd

df3 = pd.DataFrame([[key,Ingresos[key]] for key in Ingresos.keys()],
columns=["Nombre", "Salario"])
df3
```

```python
a=df3["Salario"].mean()
b=df3["Salario"].max()
c=df3["Salario"].min()
serie1 = pd.Series([a,b,c])
serie1.index = Series (["Media", "Maximo", "Minimo"])
serie1
```

EJERCICIO N° 3

```python
L1 = [["Enero", "Febrero", "Marzo", "Abril", "Marzo"],
    [30500,35600,28300,33900,42500],
    [22000, 23450, 18100, 35700, 32450]]
df1 = pd.DataFrame(L1)
df2 = df1.transpose()
df2.columns=["Mes", "Ventas", "Gastos"]
df2
```

```python
import numpy as np
df2["DC1"] =df2["Gastos"]/df2["Ventas"]*100
df2["DC2"] = np.where(df2["Ventas"] >= 3000, "Meta superada", "Meta no superada")
df2["DC3"] = np.where(df2["Gastos"] > df2["Ventas"],df2["Gastos"]-df2["Ventas"], "0")
df2["DC4"]=df2.iloc[2:5, 1].sum()
df2["DC5"]=df2.iloc[1:4, 2].sum()
df2
```
